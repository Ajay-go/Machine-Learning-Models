```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Salary_Data.csv')
df.head()
```

```python
df.describe()
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
X = df[['YearsExperience']]
Y = df[['Salary']]

x_tr,xtest,y_tr,ytest = train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
lr = linear_model.LinearRegression()


lr.fit(x_tr,y_tr)
```

```python
y_pred = lr.predict(xtest)
```

```python
lr.intercept_
lr.coef_

from sklearn.metrics import r2_score
```

```python
r2_score(ytest,y_pred)
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['YearsExperience'], df['Salary'], color='blue', label='Data')

# Plot the regression line: YearsExperience vs predicted Salary
plt.plot(df['YearsExperience'], lr.predict(X), color='red', label='Regression Line')
```
